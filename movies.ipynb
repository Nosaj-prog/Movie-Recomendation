### Movie Recomendation System

```python
#importing libraries
import pandas as pd #read and load data 
import numpy as np
import difflib #helps to understand word searched by the user
```

### Load data.

```python
moviesdata=pd.read_csv("movies.csv")
```

```python
print(moviesdata)
```

### Select Relevant Features For Recomendation

```python
selected_features=["genres", "keywords", "tagline", "cast", "director"]
selected_features
```

```python
#replace null values with null strings
for feature in selected_features:
    moviesdata[feature]=moviesdata[feature].fillna("")
```

```python
# combine the 5 selected features 
combined_features = moviesdata["genres"]+" "+moviesdata["keywords"]+" "+moviesdata["tagline"]+" "+moviesdata["cast"]+" "+moviesdata["director"]+" "
```

```python
print (combined_features)
```

```python
# convert text data to numerical data (vectorisation)
from sklearn. feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity # finding the similarities score
```

```python
#getting similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
print(similarity)
```

# get the movie from the user

```python
moviename=input("enter your favourite movie name : ")
```

```python
# find the closest match for the movie give by the user 
find_close_match=difflib.get_close_matches(moviename, moviesdata["title"])
```

```python
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
# find the index of the movie in the title 
index_of_the_movie=moviesdata[moviesdata.title== close_match]["index"].values[0]
print(index_of_the_movie)
```

```python
from sklearn.metrics.pairwise import cosine_similarity # finding the similarities score 
```

```python
similaritiesscore=list(enumerate(similarity[index_of_the_movie]))
print(similaritiesscore)
```

```python
len(similaritiesscore)
```

```python
# sort the movies based on the simlarity scores in descending order
sorted_similar_movies=sorted(similaritiesscore, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# print the name of similiar movies based on index 
print("movies suggested for you : \n")

i=1

for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=moviesdata[moviesdata.index==index]["title"].values[0]
    if (i < 30):
        print(i,".",title_from_index)
        i+=1
```
